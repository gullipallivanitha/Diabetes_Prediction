```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import pickle
```

```python
#Read Diabetic Dataset
Diabetic_dataset = pd.read_csv("C://Users//rgukt//Desktop//Diabetic_Predictio//diabetes.csv")
Diabetic_dataset.head()
```

```python
#Dataset shape
Diabetic_dataset.shape
```

```python
#Outcome values count
Diabetic_dataset["Outcome"].value_counts()
```

```python
Diabetic_dataset.groupby("Outcome").mean()
```

```python
#Sepearate the Data and Label
X=Diabetic_dataset.drop(columns=["Outcome"])
Y=Diabetic_dataset["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#The differnce b/w values in dataset is more we shouls standardize it
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X)
```

```python
# Save the scaler
with open("scaler.pkl", "wb") as file:
    pickle.dump(scaler, file)
print("Scaler object saved successfully as scaler.pkl")
```

```python
print(standarized_data)
```

```python
X=standarized_data
Y=Diabetic_dataset["Outcome"]
```

```python
print(X)
```

```python
print(Y)
```

```python
#Split training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
##training the support vector Machine Classifier

classifier = svm.SVC(kernel='linear', probability=True)
```

```python
classifier.fit(X_train,Y_train)
```

```python
with open("classifier.pkl", "wb") as file:
    pickle.dump(classifier, file)

print("✅ SVM Model saved successfully as classifier.pkl")


```

```python
import os
print(os.getcwd())  # Prints the current working directory
```

```python
# accuracy score on the training data
training_prediction=classifier.predict(X_train)
training_predict_accuracy=accuracy_score(training_prediction,Y_train)
```

```python
print("Accuracy score of the training data :", training_predict_accuracy)
```

```python
# accuracy score on the testing data
test_prediction=classifier.predict(X_test)
test_predict_accuracy=accuracy_score(test_prediction,Y_test)
```

```python
print("Accuracy score of the testing data :",test_predict_accuracy)
```
